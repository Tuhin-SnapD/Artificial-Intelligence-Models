# CV2
The code uses several OpenCV functions to perform image processing tasks such as grayscale conversion, resizing, blurring, and thresholding. It also uses imutils library for resizing image.

```python
#Import necessary packages
import cv2
import imutils
from IPython.display import Image
```

```python
# Load the image using OpenCV
img1 = cv2.imread('coding.png')
# Display
Image(filename='coding.png') 
```

```python
# Display the original image using cv2.imshow()
cv2.imshow('Original Image', img1)

# Print image information
print("Shape: ",img1.shape)
print("Size: ",img1.size)
print("Type: ",img1.dtype)

# Wait for user input and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert the image to grayscale
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
# Display the grayscale image using cv2_imshow()
cv2.imshow("Black and White Image",img2)
# Save the grayscale image as "gray_coding.png" using "cv2.imwrite()"
cv2.imwrite('bwImage.png', img2)
# Wait for user input and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
# Display
Image(filename='bwImage.png') 
```

```python
# Resize the image to a width of 200 pixels
img = imutils.resize(img1,width=200)
# Display the resized image using cv2_imshow()
cv2.imshow("Resized Image",img)
# Save the resized image as "resizedImage.png"
cv2.imwrite("resizedImage.png", img)
# Wait for user input and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
# Display
Image(filename='resizedImage.png') 
```

```python
# Apply Gaussian blur to the image
img = cv2.GaussianBlur(img1,(21,21),0)
# Display the blurred image using cv2_imshow()
cv2.imshow("Blurred Image",img)
# Save the blurred image as "gaussianImage.png"
cv2.imwrite("gaussianImage.png", img)
# Wait for user input and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
# Display
Image(filename='gaussianImage.png') 
```

```python
# Apply thresholding to the b/w image
img = cv2.threshold(img2,180,180,cv2.THRESH_BINARY)[1]
# Display the thresholded image using cv2_imshow()
cv2.imshow("Threshold Image",img)
# Save the thresholded image as "thresholdImage.jpg"
cv2.imwrite("thresholdImage.png", img)
# Wait for user input and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
# Display
Image(filename='thresholdImage.png')
```

```python
# Open the default camera
cam = cv2.VideoCapture(0)
```

```python
while True:
        # Capture a frame from the camera
        ret, frame = cam.read()

        # Check if the frame was successfully captured
        if not ret:
            print("Error: Failed to capture frame from camera")
            break

        # Display the frame in a window called "camera"
        cv2.imshow("camera", frame)

        # Wait for a key press and check if the "q" key was pressed
        key = cv2.waitKey(1) & 0xFF
        if key == ord("q"):
            break
```

```python
# Release the camera and close all windows
cv2.destroyAllWindows()
```
